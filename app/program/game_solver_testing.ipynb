```python
import csv
```

```python
def test_sudoku_solving(num_games=100):
    # Define the difficulties to test
    difficulties = ['easy', 'medium', 'hard']
    sudoku_data= []
    
    # Loop through each difficulty level
    for difficulty in difficulties:
        print(f"\nTesting difficulty: {difficulty.upper()}")
        
        
        # Generate and solve num_games Sudoku games per difficulty
        for _ in range(num_games):
            game = SudokuGame(difficulty=difficulty)
            puzzle = game.generate_game()  # Generate the game board
            
            print("Generated Sudoku Puzzle:")
            generated_puzzle_str = "\n".join(" ".join(str(val) if val != 0 else "." for val in row) for row in puzzle)
            print(generated_puzzle_str)
            
            solver = UnifiedSolver(puzzle)
            if solver.solve():
                print("Sudoku Solved! Solution:")
                solved_puzzle_str = "\n".join(" ".join(str(val) for val in row) for row in solver.board)
                print(solved_puzzle_str)
            else:
                solved_puzzle_str = "Failed to solve the Sudoku."
                print(solved_puzzle_str)
        
            sudoku_data.append([generated_puzzle_str, solved_puzzle_str, difficulty])
    
    with open('sudoku_results.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['sudoku_generated', 'sudoku_solved', 'difficulty_level'])
        writer.writerows(sudoku_data)

    return sudoku_data
```

### create running time with both solvers and compare.  

```python
import matplotlib.pyplot as plt
import numpy as np
import timeit
import csv
```

```python
def test_sudoku_solving2(num_games=1000):
    difficulties = ['easy', 'medium', 'hard']
    results2 = {'basic': {d: [] for d in difficulties}, 'advanced': {d: [] for d in difficulties}}

    for difficulty in difficulties:
        for _ in range(num_games):
            game = SudokuGame(difficulty=difficulty)
            puzzle = game.generate_game()

            solver = UnifiedSolver(puzzle)
            # Measure basic_solve time
            basic_time = timeit.timeit(solver.basic_solve, number=1)
            results2['basic'][difficulty].append(basic_time)

            # Measure advanced_solve time
            solver = UnifiedSolver(puzzle)  # Reinitialize to reset the board state
            advanced_time = timeit.timeit(solver.advanced_solve, number=1)
            results2['advanced'][difficulty].append(advanced_time)

    return results2
```

```python
def plot_results2(results2):
    fig, ax = plt.subplots(figsize=(10, 6))
    difficulties = ['easy', 'medium', 'hard']
    index = np.arange(len(difficulties))
    bar_width = 0.35

    # Calculate means
    basic_means = [np.mean(results2['basic'][d]) for d in difficulties]
    advanced_means = [np.mean(results2['advanced'][d]) for d in difficulties]

    # Plotting
    ax.bar(index - bar_width/2, basic_means, bar_width, label='Basic Solve', color='blue')
    ax.bar(index + bar_width/2, advanced_means, bar_width, label='Advanced Solve', color='orange')

    ax.set_xlabel('Difficulty')
    ax.set_ylabel('Average Time (seconds)')
    ax.set_title('Average Solving Time by Method and Difficulty')
    ax.set_xticks(index)
    ax.set_xticklabels(difficulties)
    ax.legend()

    plt.show()
```

```python
results2 = test_sudoku_solving2()
plot_results2(results2)
```

```python
#graph for total time per difficulty level for each solver 
def test_sudoku_solving3(num_games=1100):
    difficulties = ['easy', 'medium', 'hard']
    results4 = {'basic': {}, 'advanced': {}}

    for difficulty in difficulties:
        total_time_basic = 0
        total_time_advanced = 0
        
        for _ in range(num_games):
            game = SudokuGame(difficulty=difficulty)
            puzzle = game.generate_game()
            solver = UnifiedSolver(puzzle)

            # Time the execution of basic_solve for one run
            time_basic = timeit.timeit(solver.basic_solve, number=1)
            total_time_basic += time_basic  # Summing total time across all puzzles

            # Time the execution of advanced_solve for one run
            time_advanced = timeit.timeit(solver.advanced_solve, number=1)
            total_time_advanced += time_advanced  # Summing total time across all puzzles

        # Store the total time for all puzzles solved under this difficulty
        results4['basic'][difficulty] = total_time_basic
        results4['advanced'][difficulty] = total_time_advanced

    return results4

def plot_results3(results4):
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots(figsize=(10, 6))
    difficulties = ['easy', 'medium', 'hard']
    index = np.arange(len(difficulties))
    bar_width = 0.35

    basic_times = [results4['basic'][d] for d in difficulties]
    advanced_times = [results4['advanced'][d] for d in difficulties]

    ax.bar(index - bar_width/2, basic_times, bar_width, label='Basic Solve', color='blue')
    ax.bar(index + bar_width/2, advanced_times, bar_width, label='Advanced Solve', color='orange')

    ax.set_xlabel('Difficulty')
    ax.set_ylabel('Total Time (seconds)')
    ax.set_title('Total Solving Time by Method and Difficulty')
    ax.set_xticks(index)
    ax.set_xticklabels(difficulties)
    ax.legend()

    plt.show()
```

```python
#with check grid items
def test_sudoku_solving(num_games=100):
    # Define the difficulties to test
    difficulties = ['easy', 'medium', 'hard']
    results = {}
    
    # Loop through each difficulty level
    for difficulty in difficulties:
        print(f"\nTesting difficulty: {difficulty.upper()}")
        success_count = 0
        failure_count = 0
        
        # Generate and solve num_games Sudoku games per difficulty
        for _ in range(num_games):
            game = SudokuGame(difficulty=difficulty)
            puzzle = game.generate_game()  # Generate the game board
            
            print("Generated Sudoku Puzzle:")
            for row in puzzle:
                print(" ".join(str(val) if val != 0 else "." for val in row))
            
            solver = UnifiedSolver(puzzle)
            if solver.solve():
                success_count += 1
                print("Sudoku Solved! Solution:")
                for row in solver.board:
                    print(" ".join(str(val) for val in row))
                
                # Check the validity of the solution
                validity_checks = solver.check_grid_items()
                print("Grid Validity Check:")
                for row in validity_checks:
                    print(" ".join('T' if val else 'F' for val in row))
            else:
                failure_count += 1
                print("Failed to solve the Sudoku.")
        
        results[difficulty] = {'success': success_count, 'failure': failure_count}
        # Print and store results for difficulty:
        print(f"\nResults for {difficulty.upper()} difficulty:")
        print(f"Total Games: {num_games}")
        print(f"Successfully Solved: {success_count}")
        print(f"Failures: {failure_count}")
        print(f"Success Rate: {success_count / num_games * 100:.2f}%")
    
    # Print all results
    print("\nAll Test Results:")
    for difficulty, outcome in results.items():
        success_rate = outcome['success'] / (outcome['success'] + outcome['failure']) * 100
        print(f"{difficulty.capitalize()} - Successes: {outcome['success']}, Failures: {outcome['failure']}, Success Rate: {success_rate:.2f}%")

    return results
```

```python
results = test_sudoku_solving()

# Print all test results
print("\nAll Test Results:")
for difficulty, outcome in results.items():
    success_rate = outcome['success'] / (outcome['success'] + outcome['failure']) * 100
    print(f"{difficulty.capitalize()} - Successes: {outcome['success']}, Failures: {outcome['failure']}, Success Rate: {success_rate:.2f}%")
```
