```python
def test_sudoku_solving(num_games=100):
    # Define the difficulties to test
    difficulties = ['easy', 'medium', 'hard']
    sudoku_data= []
    
    # Loop through each difficulty level
    for difficulty in difficulties:
        print(f"\nTesting difficulty: {difficulty.upper()}")
        
        
        # Generate and solve num_games Sudoku games per difficulty
        for _ in range(num_games):
            game = SudokuGame(difficulty=difficulty)
            puzzle = game.generate_game()  # Generate the game board
            
            print("Generated Sudoku Puzzle:")
            generated_puzzle_str = "\n".join(" ".join(str(val) if val != 0 else "." for val in row) for row in puzzle)
            print(generated_puzzle_str)
            
            solver = UnifiedSolver(puzzle)
            if solver.solve():
                print("Sudoku Solved! Solution:")
                solved_puzzle_str = "\n".join(" ".join(str(val) for val in row) for row in solver.board)
                print(solved_puzzle_str)
            else:
                solved_puzzle_str = "Failed to solve the Sudoku."
                print(solved_puzzle_str)
        
            sudoku_data.append([generated_puzzle_str, solved_puzzle_str, difficulty])
    
    with open('sudoku_results.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['sudoku_generated', 'sudoku_solved', 'difficulty_level'])
        writer.writerows(sudoku_data)

    return sudoku_data
```

```python
#with check grid items
def test_sudoku_solving(num_games=100):
    # Define the difficulties to test
    difficulties = ['easy', 'medium', 'hard']
    results = {}
    
    # Loop through each difficulty level
    for difficulty in difficulties:
        print(f"\nTesting difficulty: {difficulty.upper()}")
        success_count = 0
        failure_count = 0
        
        # Generate and solve num_games Sudoku games per difficulty
        for _ in range(num_games):
            game = SudokuGame(difficulty=difficulty)
            puzzle = game.generate_game()  # Generate the game board
            
            print("Generated Sudoku Puzzle:")
            for row in puzzle:
                print(" ".join(str(val) if val != 0 else "." for val in row))
            
            solver = UnifiedSolver(puzzle)
            if solver.solve():
                success_count += 1
                print("Sudoku Solved! Solution:")
                for row in solver.board:
                    print(" ".join(str(val) for val in row))
                
                # Check the validity of the solution
                validity_checks = solver.check_grid_items()
                print("Grid Validity Check:")
                for row in validity_checks:
                    print(" ".join('T' if val else 'F' for val in row))
            else:
                failure_count += 1
                print("Failed to solve the Sudoku.")
        
        results[difficulty] = {'success': success_count, 'failure': failure_count}
        # Print and store results for difficulty:
        print(f"\nResults for {difficulty.upper()} difficulty:")
        print(f"Total Games: {num_games}")
        print(f"Successfully Solved: {success_count}")
        print(f"Failures: {failure_count}")
        print(f"Success Rate: {success_count / num_games * 100:.2f}%")
    
    # Print all results
    print("\nAll Test Results:")
    for difficulty, outcome in results.items():
        success_rate = outcome['success'] / (outcome['success'] + outcome['failure']) * 100
        print(f"{difficulty.capitalize()} - Successes: {outcome['success']}, Failures: {outcome['failure']}, Success Rate: {success_rate:.2f}%")

    return results
```

```python
import unittest
from unittest import TestSuite
from test import TestSudokuGame
from sudoku_game_v5 import SudokuGame
from Solver_experiment_unified import UnifiedSolver

class TestSudokuGame(unittest.TestCase):
    @classmethod
    def setUpClass(cls):
        # Generate games once for all tests to reduce setup time for each test
        cls.game_easy = SudokuGame(difficulty='easy')
        cls.game_easy.generate_game()
        cls.game_medium = SudokuGame(difficulty='medium')
        cls.game_medium.generate_game()
        cls.game_hard = SudokuGame(difficulty='hard')
        cls.game_hard.generate_game()
        cls.correct_sum = 45  # Sum of numbers 1 through 9 expected in each row, column, and block

    def check_unique(self, elements):
        """Helper method to check if all elements in a list are unique (ignoring zero)"""
        elements = [e for e in elements if e != 0]
        return len(elements) == len(set(elements))

    def check_board_validity(self, board):
        """Utility function to check that all rows, columns, and blocks are unique."""
        for i in range(9):
            row = [num for num in board[i] if num != 0]
            col = [board[j][i] for j in range(9) if board[j][i] != 0]
            start_row, start_col = 3 * (i // 3), 3 * (i % 3)
            block = [board[r][c] for r in range(start_row, start_row + 3) for c in range(start_col, start_col + 3) if board[r][c] != 0]

            assert len(set(row)) == len(row), f"Duplicate in row {i}"
            assert len(set(col)) == len(col), f"Duplicate in column {i}"
            assert len(set(block)) == len(block), f"Duplicate in block starting at {start_row},{start_col}"

    def test_full_board_validity(self):
        """Test full board validity for easy difficulty."""
        self.check_board_validity(self.game_easy.generator.board)

    def test_sums_of_rows_cols_blocks(self):
        """Test sums of rows, columns, and blocks for medium difficulty."""
        board = self.game_medium.generator.board
        for index in range(9):
            row = board[index]
            col = [board[row][index] for row in range(9)]
            start_row, start_col = 3 * (index // 3), 3 * (index % 3)
            block = [board[start_row + i][start_col + j] for i in range(3) for j in range(3)]

            self.assertEqual(sum(row), self.correct_sum, "Row does not sum to correct value")
            self.assertEqual(sum(col), self.correct_sum, "Column does not sum to correct value")
            self.assertEqual(sum(block), self.correct_sum, "Block does not sum to correct value")

    def test_puzzle_validity(self):
        """Test puzzle validity and solvability for hard difficulty."""
        self.check_board_validity(self.game_hard.generator.board)
        solver = UnifiedSolver(self.game_hard.generator.board)
        self.assertTrue(solver.solve(), "Generated puzzle is unsolvable")

    def test_solution_validity(self):
        """Test solution validity across all difficulties."""
        for game in [self.game_easy, self.game_medium, self.game_hard]:
            solver = UnifiedSolver(game.generator.board)
            self.assertTrue(solver.solve(), f"Solver failed for {game.difficulty} difficulty")
            self.assertTrue(self.verify_solution(solver.board), "Invalid solution for a sudoku puzzle.")

    def verify_solution(self, board):
        """Check if the solution is valid by ensuring each row, column, and block sums up to the correct sum."""
        for i in range(9):
            row = board[i]
            col = [board[j][i] for j in range(9)]
            start_row, start_col = 3 * (i // 3), 3 * (i % 3)
            block = [board[start_row + r][start_col + c] for r in range(3) for c in range(3)]

            if not (sum(row) == sum(col) == sum(block) == self.correct_sum):
                return False
        return True

# To run the tests
if __name__ == "__main__":
    unittest.main()
```

```python
import unittest
import sys

class ImmediateTestResult(unittest.TextTestResult):
    def printErrors(self):
        """Print the errors as they happen."""
        super().printErrors()
        # Ensure everything is flushed to stdout
        sys.stdout.flush()
    
    def addError(self, test, err):
        super().addError(test, err)
        self.printErrors()

    def addFailure(self, test, err):
        super().addFailure(test, err)
        self.printErrors()

    def addSuccess(self, test):
        super().addSuccess(test)
        print(f"OK: {test}", flush=True)

class ImmediateTestRunner(unittest.TextTestRunner):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        # Use the custom result class
        self.resultclass = ImmediateTestResult

class TestSudokuGame(unittest.TestCase):
    @classmethod
    def setUpClass(cls):
        # Expensive setup operations, run only once
        cls.puzzles = {}
        difficulties = ['easy', 'medium', 'hard']
        for difficulty in difficulties:
            game = SudokuGame(difficulty=difficulty)
            game.generate_game()
            cls.puzzles[difficulty] = game.generator.board

    def test_full_board_validity(self):
        """Test that the full board generated is valid."""
        for difficulty, board in self.puzzles.items():
            print(f"Testing full board validity for {difficulty} difficulty.")
            self.assertTrue(self.check_full_board_validity(board), f"Full board validity test failed for {difficulty}")

    def check_full_board_validity(self, board):
        # Implement the check logic here
        return True

    def test_solver_efficiency(self):
        """Test that the solver solves puzzles efficiently."""
        for difficulty, board in self.puzzles.items():
            print(f"Testing solver efficiency for {difficulty} difficulty.")
            solver = UnifiedSolver(board)
            self.assertTrue(solver.solve(), f"Solver failed for {difficulty} puzzle")

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False, testRunner=ImmediateTestRunner(verbosity=2))
```
